```python
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product
```

```python
DOM_hourly = pd.read_csv('DOM_hourly.csv',',', index_col=['Datetime'], parse_dates=['Datetime'], dayfirst=True)
DOM_hourly.head()
plt.figure(figsize(15,7))
DOM_hourly.DOM_MW.plot()
plt.ylabel('Megawatt Energy Consumption')
pylab.show()
```

Сезонность

```python
plt.figure(figsize(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(DOM_hourly.DOM_MW[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(DOM_hourly.DOM_MW[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
```

```python
import openpyxl as op;
brent = op.load_workbook(r'C:\Users\arina.borisova\Desktop\AD\2 семестр\Datasets\brent1.xlsx');
 
sheet = brent.get_sheet_by_name('нефть brent');
from itertools import islice
data = sheet.values
cols = next(data)[1:]
data = list(data)
idx = [r[0] for r in data]
data = (islice(r, 1, None) for r in data)
dfbrent = pd.DataFrame(data, index=idx, columns=cols)
```

```python
dfbrent.head()
plt.figure(figsize(15,7))
dfbrent['Value'].plot()
plt.ylabel('Brent')
pylab.show()
```

Не наблюдается ни тренда, ни сезонности, ни цикличности.

Ресемплинг

```python
dfbrent_monthly = dfbrent.resample('M').mean()
plt.figure(figsize=(12, 3))
plt.plot(dfbrent_monthly['Value'], "-o")
```
